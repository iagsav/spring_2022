```python
#8 Метод K ближайших соседей

#------------------------------------------------------------------------------------------------------------------

#Алгоритм работы таков: 
#1.Вычисление расстояния до каждого из объектов обучающей выборки 
#2.Отобрать ближайшие объекты в обучающей выборке 
#3.Класс класифицируемого объекта - это наиболее часто встречающийся класс среди К соседей
#Классифицируемый объект относится к тому классу к которому принадлежат ближайшие соседи

#------------------------------------------------------------------------------------------------------------------

from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize

file.drop(file.columns[[0]], axis = 1, inplace = True)
x = file.drop('Occupancy', axis=1)
y = file['Occupancy']

n_classes = 2
#Получаем количество классов

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
#Делим данные на обучаущие и тестовые
  
classificator = KNeighborsClassifier() 
#Указываем классификатор(К ближайших соседей)
parametrs = {'n_neighbors': range(3,30,1)}; 
#Задаем количество соседей в виде range для подбора лучшего через поиск по сетке

grid = GridSearchCV(classificator, parametrs, cv=10)
#------------------------------------------------------------------------------------------------------------------
#GridSearchCv -  находит наилучшие параметры, путем обычного перебора  
#Задаем количество блоков для перекрестной проверки и параметры для поиска по сетке
#------------------------------------------------------------------------------------------------------------------
grid.fit(x_train, y_train)
#Обучаем алгоритм
grid.best_params_
#Выводим лучшие параметры
```

```python
grid.predict(x_test)
#Делаем прогноз на тестовых данных и выводим результат
```

```python
#PR для (KNN)
from sklearn.metrics import precision_recall_curve

y_score = grid.predict_proba(x_test) 
#Делаем прогноз в виде вероятностей

precision, recall, _ = precision_recall_curve(y_test, y_score[:,1]) 

#------------------------------------------------------------------------------------------------------------------

#Для каждого класса рассчитываем precision и recall
#Формула Precision =  TP/(TP+FP)
#Формула Recall = TP/(TP+FN)
#TP - true positive 
#FP - Falce positive
#TN - true negative
#FN - falce negative 

#------------------------------------------------------------------------------------------------------------------

plt.plot(recall, precision, linestyle='--',color='green') 
#Выводим график

plt.title('PR диаграмма (KNN)')
plt.xlabel('Recall')
plt.ylabel('Precission')
plt.show()
```

```python
#ROCС
from sklearn.metrics import roc_curve

fpr, tpr, _ = roc_curve(y_test, y_score[:,1]) 
#Для каждого класса рассчитываем False-Positive rate и True-Positive rate

#------------------------------------------------------------------------------------------------------------------

#Формула false-positive rete=  FP/(FP+TN) - спецефичность алгоритма
#Формула True-Positive rate = TP/(TP+FN) - чувствительность алгоритма 
#TP - true positive 
#FP - Falce positive
#TN - true negative
#FN - falce negative 

#------------------------------------------------------------------------------------------------------------------

plt.plot(fpr, tpr, linestyle='--', color='red')
#Выводим график 

plt.plot([0, 1], [0, 1], color='green', linestyle='-')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic Curve')
plt.show()
```

```python
#9 логический классификатор (бинарное решающее дерево).

#------------------------------------------------------------------------------------------------------------------
#Основной алгоритм посторения таков:
#1. Корень дерева - вся выборка 
#2. Проверям выполение критерия останова. Если этот критерий выполняется, то нужно 
#   выбрать для узла выдаваемый прогноз
#3. Если не выпоняется, то разбить множество на несколько непересекающихся 
#
#Критерии останова: 
#
#Ограничение максимальной глубины дерева.
#Ограничение минимального числа объектов в листе.
#Ограничение максимального количества листьев в дереве.
#Останов в случае, если все объекты в вершине относятся к одному классу.
#------------------------------------------------------------------------------------------------------------------
from sklearn import tree
classificator = tree.DecisionTreeClassifier() 
#Указываем классификатор

parametrs = {'criterion': ['gini', 'entropy'], 'max_depth': range(3,30,1)};
#Указываем параметры для подбора лучшего через поиск по сетке
#criterium критерий ветвления
#Оценивает важность функции. Основываясь на нём, модель определит важность для каждого признака классификации
#Максимальная глубина дерева

grid = GridSearchCV(classificator, parametrs, cv=10)
#GridSearchCv -  находит наилучшие параметры, путем обычного перебора  
#Задаем количество блоков для перекрестной проверки и параметры для поиска по сетке

grid.fit(x_train, y_train)
#Обучаем алгоритм

grid.best_params_
#Выводим лучшие параметры
```

```python
#PR для Tree
from sklearn.metrics import precision_recall_curve

y_score = grid.predict_proba(x_test)
#Grid ищет оптимальные парамеры 
#Делаем прогноз в виде вероятностей

precision, recall, _ = precision_recall_curve(y_test, y_score[:,1]) 
#Для каждого класса рассчитываем precission и recall
#Для каждого класса рассчитываем precision и recall
#Формула Precision =  TP/(TP+FP)
#Формула Recall = TP/(TP+FN)
#TP - true positive 
#FP - Falce positive
#TN - true negative
#FN - falce negative 

plt.plot(recall, precision, linestyle='--',color='green', label='0 Type') 
#Выводим график класса 1

plt.title('PR curve (Tree)')
plt.xlabel('Recall')
plt.ylabel('Precission')
plt.show()
```

```python
#ROC
from sklearn.metrics import roc_curve

fpr, tpr, _ = roc_curve(y_test, y_score[:,1]) 
#Для каждого класса рассчитываем False-Positive rate и True-Positive rate
#Формула false-positive rete=  FP/(FP+TN) - спецефичность алгоритма
#Формула True-Positive rate = TP/(TP+FN) - чувствительность алгоритма 
#TP - true positive 
#FP - Falce positive
#TN - true negative
#FN - falce negative 

plt.plot(fpr, tpr, linestyle='--',color='red')
    
plt.plot([0, 1], [0, 1], color='green', linestyle='-')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic Curve')
plt.show()
```
