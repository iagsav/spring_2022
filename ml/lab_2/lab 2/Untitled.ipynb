```python
#1. Визуализируйте корреляции между признаками, находящимися в наборе данных (heatmap). 
plt.figure(figsize=(9, 9))

sb.heatmap(happy_df.corr(), annot=True)
```

```python
#3. Оцените количество счастья по годам.
happy_df_15 = pd.read_csv('2015.csv',sep=',')
happy_df_16 = pd.read_csv('2016.csv',sep=',')
happy_df_17 = pd.read_csv('2017.csv',sep=',')

plt.subplot(2,2,1)
plt.tight_layout()
plt.title("2015")

plt.xticks(rotation = 90, fontsize = 8);
plt.grid();

plt.plot(happy_df_15['Country'].head(25), happy_df_15['Happiness Score'].head(25))

plt.subplot(2,2,2)
plt.tight_layout()
plt.title("2016")

plt.xticks(rotation = 90, fontsize = 8);
plt.grid();

plt.plot(happy_df_16['Country'].head(25), happy_df_16['Happiness Score'].head(25))

plt.subplot(2,2,3)
plt.tight_layout()
plt.title("2017")

plt.xticks(rotation = 90, fontsize = 8);
plt.grid();

plt.plot(happy_df_17['Country'].head(25), happy_df_17['Happiness.Score'].head(25))
```

```python
#4. Определите наиболее сильно изменяющиеся параметры по разным странам в разные годы.
print("2015 year:")
std15 = happy_df_15.std(numeric_only=True)
print(std15)
print("\n")

print("2016 year:")
std16 = happy_df_16.std(numeric_only=True)
print(std16)
print("\n")

print("2017 year:")
std17 = happy_df_17.std(numeric_only=True)
print(std17)
print("\n")
```

```python
#5. Определите страны, появляющиеся и исчезающие в рейтинге стран.
sets15_16 = set(happy_df_15['Country'])
sets15_16.symmetric_difference_update(happy_df_16['Country'])

sets15_17 = set(happy_df_15['Country'])
sets15_17.symmetric_difference_update(happy_df_17['Country'])

sets15_16.union(sets15_17)
```

```python
#6. Удалите строки с пропущенными значениями. Выведите размер набора данных до и после удаления.
import random
pd.options.mode.chained_assignment = None
happy_df = pd.read_csv('2015.csv',sep=',')

print("Размер данных до удаления")
shapes = happy_df.shape
print("Количество строк: " + str(shapes[0]))

happy_df['Country'][np.random.choice(random.randint(10,shapes[0]), random.randint(0,shapes[0]))] = ''#Рандомно выбираю строки и очищаю у них информацию о стране(Т.к в таблицах нет отсутствующих значений)
happy_df.replace('', np.nan, inplace=True) #Заменяем отсутствующие значения на NaN
happy_df.dropna(inplace=True)#Удаляем все строки где есть NaN

print("Размер данных после удаления")
shapes = happy_df.shape
print("Количество строк: " + str(shapes[0]))
```

```python
#7. Вычислите корреляцию Пирсона и Спирмена между признаками Trust.Government.Corruption и Family. Постройте гистограммы для указанных признаков. Оцените значение p-value.
from scipy import stats
print("Кореляция Спирмена:")
SpearmanCor = stats.spearmanr(happy_df_15['Trust (Government Corruption)'],happy_df_15['Family']) #happy_df_15[['Trust (Government Corruption)','Family']].corr(method='spearman') #corr(method='pearson')
print("Кореляция: " + str(SpearmanCor[0]) + " p-value: " + str(SpearmanCor[1]))

print("\n\n\n")

print("Кореляция Пирсона:")
PearsonCor = stats.pearsonr(happy_df_15['Trust (Government Corruption)'],happy_df_15['Family'])
print("Кореляция: " + str(PearsonCor[0]) + " p-value: " + str(PearsonCor[1]))

print("\n\n\n")

happy_df_15[['Trust (Government Corruption)','Family']].hist();
```
