# 1. Постройте графики количества осадков по годам в разных штатах. Сделайте выводы по построенным графикам.

```python
import pandas as pd # Подключаем библиотеку pandas
# Эта команда указывает, что график необходимо построить все в той же оболочке Jyputer, но он будет выглядеть как обычная картинка
%matplotlib inline  
# Библиотека для настройки графиков
import matplotlib.pyplot as plt
```

```python
# Загружаем необходимый файл
rainHistory_df = pd.read_csv('rainfall in india 1901-2015.csv', sep = ',')
print(rainHistory_df.head())
```

```python
# Представлям данные в виде таблицы
rain_data = pd.read_csv('rainfall in india 1901-2015.csv', sep = ',')
rain_data
```

```python
# Посмотрим на признаки:
rain_data.columns
```

```python
# Берем из таблицы данные, необходимые для графиков:
gr1_df = rain_data[["SUBDIVISION", "YEAR", "ANNUAL"]]
gr1_df
```

```python
# Выводим график
fig, ax = plt.subplots()
(gr1_df.groupby(by='SUBDIVISION')
 .plot(kind='bar', stacked='true',color ='b', alpha=0.1, figsize=(20, 10), legend= False, fontsize=10, ax=ax, x='YEAR', y='ANNUAL')
)
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Annual', fontsize=12)
```

По данному графику можно сделать следующие выводы:
1) Наибольшее количество осадков на штат выпало в 1933 году
2) Наименьшее количество осадков на штат выпало в 1987 году
3) В первой половине 20-ого века в Индии выпало больше осадков, чем во второй половине

# 2. Постройте графики количества осадков по месяцам по всем годам кумулятивно. Сделайте выводы по построенным графикам.

```python
#Выбор данных для графика
#Группируем осадки по годам и суммируем их по месяцам
gr2 = rain_data.groupby(by='YEAR').sum()[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 
                                          'JUL', 'AUG', 'SEP', 'OCT', 
                                          'NOV', 'DEC']]
gr2
```

```python
#Строим график
gr2.plot(kind = 'bar', stacked = 'true', figsize = (22,12))
plt.show()
```

По данному графику можно сделать следующие выводы:
1) Наибольшее колличесво осадков выпадало в июле;
2) В летнее время выпало наибольшее колличество осадков;
3) Наименьшее колличесво осадков выпало в январе и феврале;
4) Наибольшее суммарное количество осадков выпало в 1961 году;
5) Наименьшее суммарное колличество осадков выпало в 1972 году;

# 3. Постройте графики количества осадков по штатам, используйте boxplot. Сделайте выводы по построенным графикам.

boxplot - это функция библиотеки seaborn, позволяющая строить блочные диаграммы

```python
# Подключаем seaborn
import seaborn as sns
# Устанавливаем тему графика
sns.set_style('whitegrid')
# figure - это контейнер самого верхнего уровня. Та область, в которой все нарисовано
# Создаем область figure и указываем ее размер
fig = plt.figure(figsize=(18, 9))
# pip.xticks исользуется для установки текущих местоположений тиков и меток оси x
plt.xticks(rotation='vertical')
# Указываем наименования осей и передаем данные для графика
sns.boxplot(x='SUBDIVISION', y='ANNUAL', data=rain_data)
```

По данному графику можно сделать следующие выводы:
1) Осадки в штате ARUNACHAL PRADESH имеют больший диапазон по сравнению с другими штатами;
2) Самый низкий уровень осадков наблюдается в WEST RAJASTHAN с очень низкими отклонениями, по сравнению с другими штатами;
3) Самые большие отклонения (выбросы) наблюдаются в COASTAL KARNATAKA.

# 4. Определите штаты, для которых количество осадков минимально и максимально.

```python
# Сортируем штаты по суммарным осадкам за все годы наблюдения в порядке убывания (ascending = False)
precipitation = rain_data.groupby(by='SUBDIVISION').sum()[['ANNUAL']].sort_values(by='ANNUAL', ascending = False)
precipitation
```

```python
# Находим штаты с масимальным и минимальным количеством осадков:
hig = precipitation.index.values[0]
low = precipitation.index.values[-1]
print ('Наибольшее колличество осадков: ', hig)
print ('Наименьшее колличество осадков: ', low)
```
