7. Вычислите корреляцию Пирсона и Спирмена между признаками Trust.Government.Corruption и Family. Сравните полученные величины корреляции. Постройте гистограммы для указанных признаков. Оцените значение p-value.

2015 год:

```python
from scipy import stats
data = pd.read_csv('2015.csv')
x = data['Trust (Government Corruption)']
y = data['Family']
print('Пирсон:')
print (stats.pearsonr(x,y))
print('Спирмен:')
print(stats.spearmanr(data['Trust (Government Corruption)'], data['Family']))
```

```python
data['Trust (Government Corruption)'].hist(bins = 30)
```

```python
data['Family'].hist(bins = 30)
```

2016 год:

```python
data = pd.read_csv('2016.csv')
x = data['Trust (Government Corruption)']
y = data['Family']
print('Пирсон:')
print (stats.pearsonr(x,y))
print('Спирмен:')
print(stats.spearmanr(data['Trust (Government Corruption)'], data['Family']))
```

```python
data['Trust (Government Corruption)'].hist(bins = 30)
```

```python
data['Family'].hist(bins = 30)
```

```python
data = pd.read_csv('2017.csv')
x = data['Trust..Government.Corruption.']
y = data['Family']
print('Пирсон:')
print (stats.pearsonr(x,y))
print('Спирмен:')
print(stats.spearmanr(data['Trust..Government.Corruption.'], data['Family']))
```
