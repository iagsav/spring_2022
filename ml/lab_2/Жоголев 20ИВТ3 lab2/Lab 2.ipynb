**Задание 1: Постройте графики количества осадков по годам в разных штатах. Сделайте выводы по построенным графикам.**

```python
ria_graph  = ria.groupby(['SUBDIVISION'])
fig,ax = plt.subplots(figsize = (8,6))
for label, ria_new in ria_graph:
    ria_new.plot (kind = 'line', ax=ax, label = label, x = 'YEAR', y = 'ANNUAL', figsize= (20,15))
plt.show()
```

Рассмотрев полученный график, можем сделать вывод, что количество осадков в различных штатах из выборки за последние годы сильно . Благодаря графику, мы можем визуализировать полученные значения.

**Задание 2. Постройте графики количества осадков по месяцам по всем годам кумулятивно. Сделайте выводы по построенным графикам** 

```python
ria_cum = ria[['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']]
ria_cum = ria_cum.cumsum()
ria_cum.plot(figsize=(8,8))
plt.show()
```

Так как куммулятивные кривые количества осадков имеют примерно линейный закон распределения, то можем сделать вывод, что ежегодно выпадает примерно равное количество осадков.

**Задание 9.Вычислите корреляцию Пирсона и Спирмена между признаками NOV и MAR. Сравните полученные величины корреляции. Постройте гистограммы для указанных признаков. Оцените значение p-value.**

```python
print(ria[['NOV','MAR']].corr(method='pearson')) 
print(ria[['NOV', 'MAR']].corr(method='spearman'))
sb.jointplot(x='NOV', y='MAR', data=ria); 
plt.show();
spr(ria['NOV'],ria['MAR'], nan_policy = 'omit')
```

**Вывод:**
Корреляция - взаимосвязь случайных величин. Чем больше корреляция между признаками, тем больше зависимость между ними. Исходя из вычислений видно, что рассматриваемые месяцы (март и ноябрь), являются некоррелирующими, т.к. коэфициент корреляции по Пирсону = 0,008; а по Спирману = 0,14. Так как значение p-value мало, значит отвергаем нулевую гипотезу. Коэффициент корреляции между признаками очень мал, значит можно сделать вывод что корреляция слабо выражена.
