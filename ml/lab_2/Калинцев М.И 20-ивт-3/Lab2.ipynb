1.Визуализируйте корреляции между признаками, находящимися в наборе данных (heatmap). Сделайте выводы.

```python
import pandas as pd
df = pd.read_csv('2015.csv')
df
```

```python
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt


```

```python
sns.heatmap(df.corr(),  annot=True) #функция для отображения коэффициента корреляции
```

Корреляция – это мера зависимости одного показателя от другого. Коэффициент корреляции - это число, которое находится в промежутке от -1 до 1. Отражает силу и полюс взаимосвязи величин.

r=-1 означает максимальную обратную связь.
r=1 максимальную прямую связь.
r=0 означает, что никакой связи не выявлено.
