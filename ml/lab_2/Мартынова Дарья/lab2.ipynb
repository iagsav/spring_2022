Постройте графики количества осадков по годам в разных штатах. Сделайте выводы по построенным графикам.

```python
df_graph = df.groupby(['SUBDIVISION'])
fig,ax = plt.subplots(figsize=(8,6))
for label, df_new in df_graph:
    df_new.plot(kind='line', ax=ax, label=label, x='YEAR', y='ANNUAL', figsize=(20,10))
plt.show()
```
