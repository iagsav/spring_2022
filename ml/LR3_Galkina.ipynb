```python
from sklearn.metrics import precision_recall_curve, roc_curve
from sklearn.preprocessing import label_binarize
```

```python
y_score = knn_grid.predict_proba(x_test)
```

```python
# precision recall curve
precision = dict()
recall = dict()
for i in range(7):
    precision[i], recall[i], _ = precision_recall_curve(y_test, y_score[:, i], pos_label = i)
    plt.plot(recall[i], precision[i], lw=2, label='class {}'.format(i))
    
plt.xlabel("recall")
plt.ylabel("precision")
plt.legend(loc="best")
plt.title("precision vs. recall curve")
plt.show()
```

```python
y_score[:, i]
```

```python
# roc curve
fpr = dict()
tpr = dict()

for i in range(7):
    fpr[i], tpr[i], _ = roc_curve(y_test, y_score[:, i], pos_label = i)
    plt.plot(fpr[i], tpr[i], lw=2, label='class {}'.format(i))

plt.xlabel("false positive rate")
plt.ylabel("true positive rate")
plt.legend(loc="best")
plt.title("ROC curve")
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import roc_auc_score
```

```python
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(7):
    fpr[i], tpr[i], _ = roc_curve(y_test, y_score[:, i], pos_label = i)
    roc_auc[i] = auc(fpr[i], tpr[i])

fpr["micro"], tpr["micro"], _ = roc_curve(y_test, y_score[:,2], pos_label = 2)
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
```

```python
roc_auc
```

```python
y_test = y_test + 1
y_test
```

```python
plt.figure()
lw = 2
plt.plot(
    fpr[2],
    tpr[2],
    color="darkorange",
    lw=lw,
    label="ROC curve (area = %0.2f)" % roc_auc[2],
)
plt.plot([0, 1], [0, 1], color="navy", lw=lw, linestyle="--")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver operating characteristic example")
plt.legend(loc="lower right")
plt.show()
```

```python
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(7)]))
```

```python
mean_tpr = np.zeros_like(all_fpr)
for i in range(7):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
```

```python
mean_tpr /= 7

fpr["macro"] = all_fpr
tpr["macro"] = mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])
```

```python
from itertools import cycle
```

```python
plt.figure()
plt.plot(
    fpr["micro"],
    tpr["micro"],
    label="micro-average ROC curve (area = {0:0.2f})".format(roc_auc["micro"]),
    color="deeppink",
    linestyle=":",
    linewidth=4,
)

plt.plot(
    fpr["macro"],
    tpr["macro"],
    label="macro-average ROC curve (area = {0:0.2f})".format(roc_auc["macro"]),
    color="navy",
    linestyle=":",
    linewidth=4,
)

colors = cycle(["aqua", "darkorange", "cornflowerblue", "midnightblue", "mediumslateblue","darkmagenta", "mediumturquoise"])
for i, color in zip(range(7), colors):
    plt.plot(
        fpr[i],
        tpr[i],
        color=color,
        lw=lw,
        label="ROC curve of class {0} (area = {1:0.2f})".format(i, roc_auc[i]),
    )

plt.plot([0, 1], [0, 1], "k--", lw=lw)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Some extension of Receiver operating characteristic to multiclass")
plt.legend(loc="lower right")
plt.show()
```
