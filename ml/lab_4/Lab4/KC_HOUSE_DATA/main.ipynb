```python
import pandas as pd
```

```python
df = pd.read_csv('kc_house_data.csv')
df
```

```python
#удаление дубликатов

print('До удаления:', df.shape)

df.drop_duplicates(keep ='first', inplace =True)  
print('После удаления:', df.shape)
```

```python
#Замена категориальных признаков

import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

not_categorial = df.drop(['date', 'id', 'waterfront', 'view', 'condition', 'grade', 'yr_built', 'yr_renovated', 'zipcode'], axis=1)
categorial = df.drop(not_categorial, axis=1)
categorial
#в данной таблице категориальные признаки дропнуты(они есть, но не выведены)
```

```python
#превращаю категориальные признаки в числа
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

for column in categorial:
    categorial[column] = pd.Series(label_encoder.fit_transform(categorial[column]))

categorial
```

```python
#Масштабирование количественных признаков

import seaborn as sns
from sklearn.preprocessing import StandardScaler
    
normilized_data = not_categorial
sns.boxplot(data = normilized_data, orient = 'h')
print('До масштабирования:')

```

```python
scaler = StandardScaler()
scaler.fit(normilized_data)
normilized_data = scaler.transform(normilized_data)

normilized_data = pd.DataFrame(normilized_data, columns = not_categorial.columns)

print('После масштабирования:')
sns.boxplot(data = normilized_data, orient = 'h')

```

```python
df = pd.concat([normilized_data, categorial], axis = 1)
df
```

```python
corr_df = df.corr()
corr_df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 12))
sns.heatmap(corr_df)
```

```python
df.drop(['id', 'date', 'yr_renovated'], axis=1, inplace=True)
df
```

```python
#Заполнение пропущенных данных 

df.isnull().sum()
```

```python
#Линейная регрессия 
#предсказание одного параметра (Y) по известному параметру X, где X — набор параметров, характеризующий наблюдение.
#апроксимация
#научный метод, состоящий в замене одних объектов другими, в каком-то смысле близкими к исходным, но более простыми.

from sklearn.model_selection import GridSearchCV, train_test_split

x, y = df.drop('price', axis = 1), df['price']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

from sklearn.linear_model import LinearRegression

linear = LinearRegression()

linear.fit(x_train, y_train)

```

```python
linear.coef_

```

```python
from scipy.stats import pearsonr, spearmanr

predict = linear.predict(x_test)
pearson, P_p_value = pearsonr(predict, y_test)
spearman, S_p_value = spearmanr(predict, y_test)

print("Корреляция Спирмена: ", spearman)
print("Корреляция Пирсона: ", pearson)

if P_p_value < 0.05:
    print('p_value (Пирсон) достоверно')
else:
    print('p_value (Пирсон) не достоверно')
if S_p_value < 0.05:
    print('p_value (Спирмен) достоверно')
else:
    print('p_value (Спирмен) не достоверно')
```

```python
#Линейная регрессия. LASSO регуляризатор
#заключается во введении дополнительного слагаемого регуляризации в функционал оптимизации модели, что часто позволяет получать более устойчивое решение.
#Нужно для избежания переобучения
from sklearn.linear_model import Lasso
import numpy as np

las = Lasso()

params = {'alpha': np.arange(0.01, 0.1, 0.01), 'max_iter' : np.arange(100,110)}
#arange() — это одна из процедур создания массива на основе числовых диапазонов. 
#Он создает экземпляр ndarray с равномерно распределенными значениями и возвращает ссылку на него.
#start это число (целое или десятичное), которое определяет первое значение в массиве.
#stop это число, которое определяет конец массива и не включается в массив.
#step это число, которое определяет интервал (разница) между каждыми двумя последовательными значениями в массиве и по умолчанию 1.
#dtype является типом элементов выходного массива и по умолчанию None.
grid = GridSearchCV(las, params)
grid.fit(x_train, y_train)
```

```python
best_alpha = grid.best_estimator_.alpha
best_max_iter = grid.best_estimator_.max_iter
best_alpha
```

```python
las = Lasso(alpha = best_alpha, max_iter = best_max_iter)
las.fit(x_train, y_train)
pred_las = las.predict(x_test)
pred_las
```

```python
pearson, P_p_value = pearsonr(pred_las, y_test)
spearman, S_p_value = spearmanr(pred_las, y_test)

print("Корреляция Спирмена: ", spearman)
print("Корреляция Пирсона: ", pearson)

if P_p_value < 0.05:
    print('p_value (Пирсон) достоверно')
else:
    print('p_value (Пирсон) не достоверно')
if S_p_value < 0.05:
    print('p_value (Спирмен) достоверно')
else:
    print('p_value (Спирмен) не достоверно')
```
