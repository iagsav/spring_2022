Набор данных содержит 9357 экземпляров усредненных по часам ответов от массива из 5 химических датчиков оксида металла, встроенных в Мультисенсорное устройство для измерения качества воздуха. Устройство было расположено на поле в сильно загрязненном районе, на уровне дороги, в пределах итальянского города. Данные были записаны с марта 2004 года по февраль 2005 года (в течение одного года), что представляет собой самые продолжительные записи, доступные в свободном доступе о срабатывании химических датчиков качества воздуха, установленных на местах. Среднечасовые концентрации СО, неметановых углеводородов, бензола, Оксиды азота (NOx) и диоксид азота (NO2) были получены с помощью совместно расположенного эталонного сертифицированного анализатора.

Имеются свидетельства перекрестной чувствительности, а также дрейфа как концепции, так и датчиков, как описано в De Vito et al., Sens. И Act. B, Vol. 129,2,2008 (требуется цитирование), что в конечном итоге влияет на возможности оценки концентрации датчиков. Отсутствующие значения помечаются значением -200.
Этот набор данных может быть использован исключительно в исследовательских целях. Коммерческие цели полностью исключены.

Информация: Дата (ДД/ММ/ГГГГ); Время (ЧЧ.ММ.СС); Истинная почасовая усредненная концентрация CO в мг/м^3 (эталонный анализатор); PT 08.S1 (оксид олова) усредненный по часам отклик датчика (номинально ориентированный на CO); Истинная почасовая усредненная общая концентрация неметаллических углеводородов в мкг/м^3 (эталонный анализатор); Истинная почасовая усредненная концентрация бензола в мкг/м^3 (эталонный анализатор); PT08.S2 (titania) усредненный по часам отклик датчика (номинально ориентированный на NMHC); Истинная почасовая усредненная концентрация NOx в ppb (эталонный анализатор); PT08.S3 (оксид вольфрама) усредненный по часам отклик датчика (номинально нацеленный на NOx); Истинная почасовая усредненная концентрация NO2 в мкг/м^3 (эталонный анализатор); PT 08.S4 (оксид вольфрама) усредненный по часам отклик датчика (номинально НЕ нацелен на 2); PT08.S5 (оксид индия) усредненный по часам отклик датчика (номинально ориентированный на O3); Температура в °C; Относительная влажность (%); Абсолютная влажность AH

```python
from sklearn.impute import SimpleImputer
```

```python
import numpy as np
```

```python
imp = SimpleImputer(missing_values = np.NaN, strategy = "most_frequent") #заполнение значений NaN самым повторяющимся значением
```

```python
air_2_df = pd.DataFrame(imp.fit_transform(air_df), columns=air_df.columns).astype(air_df.dtypes.to_dict())
```

```python
air_2_df.columns = air_df.columns
air_2_df.index   = air_df.index
```

```python
print(air_2_df.isnull().values.any()) 
```

```python
air_2_df
```

```python
data_object = air_2_df.select_dtypes(include=[object])
data_object
```

```python
data_int = air_2_df.select_dtypes(include=[np.number])
data_int
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data_object_encoded =data_object.apply(le.fit_transform)
```

```python
data_object_encoded.reset_index(drop=True, inplace=True)
data_int.reset_index(drop=True, inplace=True)
```

```python
data_2 = pd.concat([data_int, data_object_encoded], axis = 1)
```

```python
data_2
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns;
```

```python
df = pd.DataFrame(data=data_2)
sns.boxplot(x="variable", y="value", data=pd.melt(df));
```

```python
air_df_2 = data_2   #Использование максимального абсолютного масштабирования
for column in air_df_2:
  air_df_2[column] = air_df_2[column] / air_df_2[column].abs().max()
```

```python
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
y = pd.Series(air_df_2['C6H6(GT)'])
```

```python
y
```

```python
x = pd.DataFrame(data=air_df_2, columns=['CO(GT)','PT08.S1(CO)','NMHC(GT)','PT08.S2(NMHC)','NOx(GT)','PT08.S3(NOx)','NO2(GT)','PT08.S4(NO2)','PT08.S5(O3)','T','RH','AH'])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
import numpy as np
```

```python
x.shape, y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state = 12)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.linear_model import LinearRegression #линейная регрессия без использования регуляризаторов
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_test,y_test)
```

```python
y_pred = lr.predict(x_test)
```

```python
np.mean(y_pred)
```

```python
lr.intercept_ # свободный коэффициент; то, чему равна зависимая переменная, когда предиктор равен нулю
```

```python
np.mean(lr.coef_) # среднее значение угла наклона прямой
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
from sklearn.metrics import r2_score #ошибка регрессии метрики
```

```python
r2_score(y_test, y_pred)
```

```python
df['Actual'].corr(df['Predicted'],'spearman')
```

```python
df['Actual'].corr(df['Predicted'])
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
plt.plot(y_test, y_pred, '.')
X = np.linspace(0,2,2)
Y = X
plt.plot(X,Y)
plt.show()
```

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn import metrics
```

```python
predictions = cross_val_predict(lr, x, y, cv=10)
plt.scatter(y, predictions)
```
