```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
#1 Изучить набор данных
glass_df = pd.read_csv('5_energy/energydata_complete.csv',sep=',')
glass_df
```

```python
#date Дата проведения измерений
#Appliances Потребление энергии бытовыми приборами
#lights Потребление энергии светильниками
#T1, Температура в кухонной зоне
#RH_1, Влажность в кухонной зоне
#T2, Температура в гостиной
#RH_2, Влажность в гостиной
#T3, Температура в зоне прачечной
#RH_3, Влажность в помещении прачечной
#T4, Температура в офисном помещении
#RH_4, Влажность в офисном помещении
#T5, Температура в ванной комнате
#RH_5, Влажность в ванной комнате
#T6, Температура снаружи здания
#RH_6, Влажность снаружи здания
#T7, Температура в гладильной комнате
#RH_7, Влажность в гладильном помещении
#T8, Температура в комнате детей
#RH_8, Влажность в комнате детей
#T9, Температура в комнате родителей
#RH_9, Влажность в комнате родителей
#To, Температура снаружи (от метеостанции Кьевр)
#Pressure (от метеостанции Кьевре)
#RH_out Относительная влажность воздуха снаружи (от метеостанции Кьевре)
#Wind speed (от метеостанции Кьевр)
#Visibility (с метеостанции Кьевр), в км
#Tdewpoint Точка росы(от метеостанции Кьевр)
#rv1, Случайная величина 1
#rv2, Случайная величина 2
```

```python
#2 Удалить дубликаты строк
pd.options.mode.chained_assignment = None

print("Размер данных до удаления")
shapes = glass_df.shape
print("Количество строк: " + str(shapes[0]))

glass_df.drop_duplicates(inplace=True)
print("\n")

print("Размер данных после удаления")
shapes = glass_df.shape
print("Количество строк: " + str(shapes[0]))
```

```python
# Приведение даты к Timestamp формату
date = glass_df['date']
glass_df.drop(columns='date', inplace=True)
glass_df['date'] = pd.to_datetime(date).values.tolist()
glass_df
```

```python
#3 Масштабирование
plt.subplot(1, 2, 1);
plt.title('До нормализации:', fontsize=17)
plt.boxplot(glass_df);

glass_df = (glass_df - glass_df.min())/(glass_df.max() - glass_df.min());#(X-Xmin)/(Xmax-Xmin)

plt.subplot(1, 2, 2);
plt.title('После нормализации:', fontsize=17)
plt.boxplot(glass_df);
```

```python
#5 Оценить кореляцию и удалить признаки, которые коррелируют с наибольшим числом других
plt.figure(figsize=(18, 13));

corr = glass_df.corr();

sb.heatmap(corr, annot=True);
```

```python
y = glass_df['Appliances']
glass_df.drop(columns='RH_6', inplace=True)
glass_df.drop(columns='Appliances', inplace=True)
x = glass_df
```

```python
from sklearn.model_selection import train_test_split
pd.options.display.float_format ='{:,.20f}'.format
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10) #размер тестовой выборки 30%
```

```python
UsePValue = True
if UsePValue == True:
    import statsmodels.api as sm
    x_incl_const = sm.add_constant(x_train)

    model = sm.OLS(y_train, x_incl_const)
    results = model.fit()

    pd.DataFrame({'coef': results.params, 'p-value': round(results.pvalues, 3)}) #p-value - насколько важен данный параметр
    #если >5% (0.05), то данные параметры не очень важны
    glass_df.drop(columns='T1', inplace=True)
    glass_df.drop(columns='RH_4', inplace=True)
    glass_df.drop(columns='T5', inplace=True)
    glass_df.drop(columns='RH_5', inplace=True)
    glass_df.drop(columns='T7', inplace=True)
    glass_df.drop(columns='Press_mm_hg', inplace=True)
    glass_df.drop(columns='RH_out', inplace=True)
    glass_df.drop(columns='rv1', inplace=True)
    glass_df.drop(columns='rv2', inplace=True)
    x = glass_df
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10) #размер тестовой выборки 30%
```

```python
#Линейная регрессия
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

regr = LinearRegression()

ys = y.values.reshape(-1,1)

cv = cross_val_score(regr,x,ys, cv=10)
mean_cv = np.mean(cv)
print('CV: ', mean_cv)
regr.fit(x_train, y_train) #обучаем модель на тренировочной выборке


print('Traning data r-squared:', regr.score(x_train, y_train)) #насколько точная train модель
print('Test data r-squared:', regr.score(x_test, y_test)) #насколько точная test модель

print('Intercept', regr.intercept_)
pd.DataFrame(data=regr.coef_, index=x_train.columns, columns=['coef'])
```

```python
#from sys import maxsize
y_pred = regr.predict(x_test) #прогнозы по тестовым данным
#np.set_printoptions(threshold=maxsize)
np.set_printoptions(threshold=1000)
print('predicted response:', y_pred, sep='\n')
```

```python
df_check = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred}) #реальные и предсказанные значения
df_check
```

```python
A_P = pd.DataFrame({'Accuracy': y_test - y_pred}) # реальные - предсказанные значения = точность решения задачи
A_P
```

```python
#предсказания
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
#Линейная регрессия, RIDGE регуляризатор
from sklearn.linear_model import Ridge
ridle = Ridge(alpha=0.1)

cv = cross_val_score(ridle,x,ys, cv=10)
mean_cv = np.mean(cv)
print('CV: ', mean_cv)
ridle.fit(x_train, y_train)
y_train_pred = ridle.predict(x_train)

print('Traning data r-squared:', ridle.score(x_train, y_train)) #насколько точная train модель
print('Test data r-squared:', ridle.score(x_test, y_test)) #насколько точная test модель
```

```python
#from sys import maxsize
y_test_pred = ridle.predict(x_test) #прогнозы по тестовым данным
#np.set_printoptions(threshold=maxsize)
np.set_printoptions(threshold=1000)
print('predicted response:', y_test_pred, sep='\n')
```

```python
df_check = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred}) #реальные и предсказанные значения
df_check
```
